Alpha Beta Prunning

```python
# Alpha-Beta Pruning Implementation

# Function to implement minimax with alpha-beta pruning
def alpha_beta_pruning(depth, nodeIndex, isMax, values, alpha, beta, maxDepth):
    # Base case: when we reach a leaf node
    if depth == maxDepth:
        return values[nodeIndex]

    if isMax:
        best = float('-inf')

        # Recur for left and right children
        for i in range(2):
            val = alpha_beta_pruning(depth + 1, nodeIndex * 2 + i, False, values, alpha, beta, maxDepth)
            best = max(best, val)
            alpha = max(alpha, best)

            # Alpha Beta Pruning condition
            if alpha >= beta:
                break

        return best

    else:
        best = float('inf')

        # Recur for left and right children
        for i in range(2):
            val = alpha_beta_pruning(depth + 1, nodeIndex * 2 + i, True, values, alpha, beta, maxDepth)
            best = min(best, val)
            beta = min(beta, best)

            # Alpha Beta Pruning condition
            if alpha >= beta:
                break

        return best


# -----------------------------
# EDIT VALUES HERE
# -----------------------------
# Leaf node values (left to right)
values = [8, -2, 4, 7, 3, 6, 1, 0]   # Change this list as you want

# Depth of tree = log2(number of leaf nodes)
maxDepth = 3  # For 8 values => depth 3

# Initial alpha and beta values
alpha = float('-inf')
beta = float('inf')

# Root node call
optimal_value = alpha_beta_pruning(0, 0, True, values, alpha, beta, maxDepth)

print("The optimal value is:", optimal_value)
```
